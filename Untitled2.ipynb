### 1.tf.Variable相关属性

```python
import tensorflow as tf
import numpy as np

v1 = tf.Variable(tf.constant(2.0, shape=[4]), dtype='float32')
print(v1)

v2 = tf.Variable(np.ones(shape=[4,3]), dtype='float32')
print(v2)

v3 = tf.Variable(tf.keras.initializers.RandomNormal()(shape=[3,4,5]), dtype='float32')
print(v3)
```

### 2.tf.Tensor

产生张量

```python
v = tf.Variable(np.ones(shape=[4,3]), dtype='float32')
b = v * 3.0
print(b)
print(type(b).__name__)
```

 张量相加

```python
a = tf.constant(2, shape=[4], dtype='float32')
b = tf.constant(3, shape=[4], dtype='float32')
c = tf.add(a,b)
print(c)
assert isinstance(c, tf.Tensor)
```

### 3.tf.Operation

张量乘法、加法、张量比较等

```python
import tensorflow as tf
import numpy as np
a = tf.constant(4, shape=[4], dtype='float32') 
b = tf.constant(2, shape=[4], dtype='float32')

print(a)
print(b)
c = a+b
print(c)

e = a*b
print(e)

a = tf.constant([[1,2,3],[4,5,6]])
b = tf.constant([[5,4,3],[3,2,1]])

equal_check = (a==b)
print(equal_check)

d = tf.constant([[1,2,3],[3,4,5],[6,5,4]]) 
d_max1 = tf.argmax(d, axis=0)
print(equal_check)

d = tf.constant([[1,2,3],[3,4,5],[6,5,4]]) 
d_min1 = tf.argmin(d, axis=1)
print(d_min1)


e = tf.constant([1,2,3,4,5])
e_cumsum = tf.cumsum(e)
print(e_cumsum)
```

### 4.矩阵乘法

```python
from PIL import Image
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

# Loading the image and converting to a TensorFlow tensor
x_rgb = np.array(Image.open("D:/桌面/baboon.jpg")).astype('float32')
x_rgb = tf.constant(x_rgb)

# Defining the matrix used to convert the image to grayscale
grays = tf.constant([[0.3], [0.59] ,[0.11]])

# Converting the image to grayscale
x = tf.matmul(x_rgb, grays)
x = tf.squeeze(x)
print("The size of the final image: {}".format(x.shape))

# Plotting the image
f, axes = plt.subplots(1,1, figsize=(5,5))

axes.imshow(x.numpy(),cmap='gray')
axes.axis('off')
```

```python
from PIL import Image
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

# 加载图像并转换为TensorFlow张量
img_path = "D:/桌面/baboon.jpg"
x_rgb_np = np.array(Image.open(img_path)).astype('float32')  # shape: (H, W, 3)
x_rgb = tf.constant(x_rgb_np)
print(f"原始RGB图像shape: {x_rgb.shape}")

# 2. 定义灰度转换矩阵（RGB权重）
grays = tf.constant([[0.3], [0.59], [0.11]])  # shape: (3, 1)

# 3. 灰度转换（修复维度不匹配问题）
H, W, C = x_rgb.shape
x_rgb_flat = tf.reshape(x_rgb, [-1, C])  # 展平为(H*W, 3)，适配matmul维度
x_gray_flat = tf.matmul(x_rgb_flat, grays)  # 矩阵乘法: (H*W,3) @ (3,1) = (H*W,1)
x_gray = tf.reshape(x_gray_flat, [H, W])  # 还原为(H, W)
x_gray = tf.squeeze(x_gray)  # 去除多余维度（可选，已为2维）
print(f"灰度图像shape: {x_gray.shape}")

# 4. 一行展示RGB图和灰度图
f, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1行2列，调整画布大小

# 显示原始RGB图
axes[0].imshow(x_rgb_np / 255.0)  # 归一化到0-1更适配imshow（可选，0-255也可）
axes[0].set_title("Original RGB Image", fontsize=12)
axes[0].axis('off')

# 显示灰度图
axes[1].imshow(x_gray.numpy(), cmap='gray')
axes[1].set_title("Grayscale Image", fontsize=12)
axes[1].axis('off')

# 调整子图间距，避免重叠
plt.tight_layout()
plt.show()
```
